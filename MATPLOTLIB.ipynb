# A PLOT WITH ONLY Y LIST. A GRAPH WITHOUT GRID.

```
import matplotlib.pyplot as p
p.plot([2,4,6,4])
p.xlabel("Indices")
p.ylabel("Numbers")
p.title("MY PLOT")
p.show()
```

# A GRAPH WITH GRID

```
import matplotlib.pyplot as p1
p1.plot([1,2,3,4],[1,4,9,16])
p1.xlabel("Numbers")
p1.ylabel("Square of  a number")
p1.title("SQUARES")
p1.grid()
p1.show()
```

# A GRAPH WITH SQUARE DOTS

```
import matplotlib.pyplot as p1
p1.plot([1,2,3,4],[1,4,9,16],'ro')
p1.xlabel("Numbers")
p1.ylabel("Square of  a number")
p1.title("SQUARES")
p1.grid()
p1.show()
```

# A GRAPH WITH MULTIPLE LABELS

```
import matplotlib.pyplot as p2
import numpy as np
s=np.arange(.0,5.,.2)
p2.plot(s,s**2,'b--',label="^2")
p2.plot(s,s**2.2,'rs',label="^2.2")
p2.plot(s,s**2.5,'g^',label="^2.5")
p2.grid()
p2.legend()
p2.show()
```

# A GRAPH WITH INCREASING LINEWIDTH

```
import matplotlib.pyplot as p1
p1.plot([1,2,3,4],[1,4,9,16],linewidth=10.0)
p1.xlabel("Numbers")
p1.ylabel("Square of  a number")
p1.title("SQUARES")
p1.grid()
p1.show()
```

```
import matplotlib.pyplot as plt
x1=[1,2,3,4]
y1=[1,4,9,16]

x2=[1,2,3,4]
y2=[2,4,6,8]

line=plt.plot(x1,y1,x2,y2)

plt.setp(line[0],color='r',linewidth=4.0,label="Squares")
plt.setp(line[1],color='g',linewidth=2.0,label="Multiple of 2")
plt.grid()
plt.legend()
plt.show()
```

# A GRAPH WITH A FIGURE HAVING 2 SUBPLOTS

```
import matplotlib.pyplot as p11
import numpy as n
def f(t):
  return n.exp(-t)*n.cos(2*n.pi*t)
t1=n.arange(.0,5.0,0.1)
t2=n.arange(.0,5.0,0.02)

p11.figure(1)
p11.subplot(211)
p11.plot(t1,f(t1),'b--')

p11.subplot(212)
p11.plot(t2,n.cos(2*n.pi*t2),'r--')
p11.show()
```

# A GRAPH WITH MULTIPLE FIGURES

```
import matplotlib.pyplot as pp
pp.figure(1)
pp.subplot(211)
pp.plot([1,2,3])
pp.subplot(212)
pp.plot([4,5,6])

pp.figure(2)
pp.plot([1,2,4])

pp.figure(1)
pp.title("Easy as 1,2,3")
pp.show()
```
